```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams.update({'errorbar.capsize':8})
```

```python
mean_x = [2**i for i in range(1,9)]
mean_y = [0.889412, 0.370032, 0.215923, 0.203065, 0.195257, 0.193056, 0.192537, 0.157416]
std_y  = [0.001291466005919, 0.000139028374402, 0.000152173146996, 3.56682242650586e-05, \
          5.85092585266091e-05, 1.83787316694536e-05, 2.35937844922481e-05, 0.010519320003372]

fig = plt.figure(figsize=(18,12))
plt.style.use('seaborn-darkgrid')
plt.errorbar(mean_x, mean_y, fmt='b--o', yerr=std_y, ecolor='r', elinewidth=1.5, capsize=8)
plt.title('Time measurements for img_size=16384x16384, f_radius=32')
plt.xlabel('Block_size')
plt.ylabel('Time (s)')
plt.show()
```
